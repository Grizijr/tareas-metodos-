```python
import numpy as np

#primer punto encontrar el numero de acondicionamiento 

def f(x):
    return (1 - np.cos(x)) / np.sin(x)

def df(x):
     return (np.sin(x)**2 - (1 - np.cos(x)) * np.cos(x)) / (np.sin(x)**2)

def condicionamiento (x):
    return abs(x * df(x) / f(x))

#pruebo con x=1e-6

x = 1e-6
print("punto 1")

print("f(x) =", f(x))
print("f'(x) =", df(x))
print("numero condicionamiento  =", condicionamiento(x))

#segundo punto calcular f(1e-6) usando cadena de 4 operaciones elementales 

print("punto 2")

#operaciones elementales

a = np.cos(x)    #coseno
b = 1 - a        #resta
c = np.sin(x)    #seno
d = b/c          #division

#condicionamiento

k_a = abs(-x*np.sin(x)/np.cos(x))
k_b = (abs(1) + abs(a)) / abs(1 - a)
k_sin = abs(x*np.cos(x)/np.sin(x))

def f(x): return (1 - np.cos(x))/np.sin(x)
def f_prime(x, h=1e-8): return (f(x+h)-f(x-h))/(2*h)
k_f = abs(x * f_prime(x) / f(x))


print("Condicionamiento cos =", k_a)
print("Condicionamiento resta =", k_b)
print("Condicionamiento sin =", k_sin)
print("Condicionamiento division =", k_f)


#tercer punto calcular g(x)

print("punto 3")

def g(x):
    return 2 * np.sin(x/2)**2 / np.sin(x)

def g_prime(x, h=1e-8):
    return (g(x+h) - g(x-h)) / (2*h)

k_g = abs(x * g_prime(x) / g(x))

#operaciones elementales

r = x / 2          #division dentro del seno
s = np.sin(r)      #seno 1
t = s * s          #potencia
u = 2 * t          #multiplicacion por una constante
v = np.sin(x)      #seno2
w = u / v          #division total

#condicionamiento 

k_r = abs(x * (1/2) / r)
k_s = abs(r * np.cos(r) / s)  
k_t = 2.0                         
k_u = 1.0
k_v = abs(x * np.cos(x) / v)
k_W = (abs(u) + abs(v)) / abs(w)  


print("Condicionamiento division =", k_r)
print("Condicionamiento seno 1 =", k_s)
print("Condicionamiento potencia =", k_t)
print("Condicionamiento multiplicacion =", k_u)
print("Condicionamiento seno 2 =", k_v)
print("Condicionamiento division total =", k_W)
print("Condicionamiento g final =", k_g)


#ultimo punto comparar resultados y ver cual es el mas precisio

print("punto 4")

#valores reales aproximados con serie de Taylor 
f_real = x / 2
g_real = x / 2

#valores normales
f_calc = f(x)
g_calc = g(x)

#errores relativos
err_f = abs((f_calc - f_real) / f_real)
err_g = abs((g_calc - g_real) / g_real)

print("f(x) =", f_calc, " error relativo =", err_f)
print("g(x) =", g_calc, " error relativo =", err_g)

if err_f < err_g:
    print("Conclusión: f(x) es más precisa en este caso.")
else:
    print("Conclusión: g(x) es más precisa en este caso.")
```
